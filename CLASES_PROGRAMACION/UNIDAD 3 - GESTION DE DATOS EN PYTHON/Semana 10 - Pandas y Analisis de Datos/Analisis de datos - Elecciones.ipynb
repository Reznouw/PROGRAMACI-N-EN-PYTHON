## Resultados de la segunda vuelta

```python
df = pd.read_csv("Resultados_2da_vuelta.csv", encoding='latin-1', delimiter=';')
df.head()
```

### Inspección de los datos

```python
df.info()
```

```python
df.rename(columns={'VOTOS_P1': 'Peru Libre', 
                   'VOTOS_P2': 'Fuerza Popular', 
                   'VOTOS_VB': 'Votos Blanco', 
                   'VOTOS_VN': 'Votos Nulos', 
                   'VOTOS_VI': 'Votos Invalidos'}, inplace=True)
```

```python
departamentos = ['AMERICA', 'OCEANIA', 'AFRICA', 'ASIA', 'EUROPA']
df_filter = df['DEPARTAMENTO'].isin(departamentos) != True
votacion_nac = df[df_filter]
```

```python
votacion_nac.head()
```

```python
votacion_nac.info()
```

```python
votacion_nac.describe()
```

```python
votacion_nac = votacion_nac.fillna(0)      # inplace=True genera un SettingWithCopyWarning!
```

```python
votacion_nac.head()
```

### Distribución de la votación

```python
votacion_nac['Votos_BI'] = votacion_nac.loc[:, ['Votos Nulos', 'Votos Blanco', 'Votos Invalidos']].sum(axis=1)
votacion_nac.drop(columns=['Votos Nulos', 'Votos Blanco', 'Votos Invalidos'], inplace=True)
```

```python
votacion_nac.head()
```

```python
votacion_nac.tail()
```

```python
votacion_nac.loc[:,['Peru Libre', 'Fuerza Popular', 'Votos_BI']].sum()
```

```python
votacion_nac_result = votacion_nac.loc[:,['Peru Libre', 'Fuerza Popular', 'Votos_BI']].sum()
votacion_nac_result.name = "Votacion"
votacion_nac_result.plot(kind='pie', figsize=(10, 6), title='Resultados 2da vuelta', autopct="%.1f%%")
plt.show()
```

### Distribucion de los candidatos por regiones

```python
votacion_nac[['DEPARTAMENTO', 'Peru Libre', 'Fuerza Popular']].groupby(by='DEPARTAMENTO').sum()
```

```python
votacion_dep = votacion_nac[['DEPARTAMENTO', 'Peru Libre', 'Fuerza Popular']].groupby(by='DEPARTAMENTO').sum()
votacion_dep.plot(kind='bar', figsize=(12, 8), logy=True, 
                  title="Votacion por Departamento", ylabel='Votos (log scale)')
plt.show()
```

### De forma tabulada, en que departamentos ganaron los candidatos

```python
df_filter = votacion_dep['Peru Libre'] > votacion_dep['Fuerza Popular']
dep_peru_libre = votacion_dep[df_filter]
dep_peru_libre
```

```python
df_filter = votacion_dep['Peru Libre'] < votacion_dep['Fuerza Popular']
dep_fuerza_popular = votacion_dep[df_filter]
dep_fuerza_popular
```

```python
print(f"Departamento Peru Libre: {len(dep_peru_libre)}")
print(f"Departamento Fuerza Popular: {len(dep_fuerza_popular)}")
```

### ipwidgets: Votos por Candidato por departamento

```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import display

df = pd.read_csv("Resultados_2da_vuelta.csv", encoding='latin-1', delimiter=';')
df.rename(columns={'VOTOS_P1': 'PERU LIBRE', 
                   'VOTOS_P2': 'FUERZA POPULAR'}, inplace=True)
datos = df[['DEPARTAMENTO', 'PERU LIBRE', 'FUERZA POPULAR']].groupby(by="DEPARTAMENTO").sum()

lista_departamentos = list(datos.index)

@interact(Lugar=widgets.Dropdown(options=lista_departamentos))
def graph_results(Lugar):
    data_out = datos.filter(items=[Lugar], axis=0)
    ax = data_out.plot(kind='bar', figsize=(8, 6))
    ax.set_xlabel('')
    ax.set_xticks([])
    return data_out
```
